```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = np.genfromtxt("a15-spline-input.txt").T
x_new, y_new = np.genfromtxt("a15-spline-output.txt").T
```

```python
x.shape
```

```python
x_new.shape
```

```python
plt.plot(x, y, '.', label='original data', color='g', ms=10)
plt.plot(x_new, y_new, '.', label='interpolated data', color='b', ms=1)

plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend()

plt.show()
```
